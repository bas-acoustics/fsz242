# The Which Blair Test

## Introduction

The “Which Blair Project” provides a quick visual test for evaluating colour maps
(Rogowitz & Kalvin 2001). In psychophysical testing, the perceptual quality of colour
maps was assessed by using them to render a photograph of a human face. A strong
correlation was found between the perceived naturalness of images, luminance monotonicity
and the suitability of colour maps for rendering continuous scalar data.


## Method

```julia
using Images
using ColorSchemes
```

The image of Tony Blair, adapted for our experiments, is courtesy of the World Affairs
Council and is licensed under the terms of the Creative Commons Attribution 2.0 Generic
license. (We don't have the image from the original paper).

```julia
img = load("data/processed/blair/blair.png");
```

```julia
imresize(img,ratio=0.2)
```

```julia
function which_blair_test(colorscheme::ColorScheme)
    k = length(colorscheme.colors)
    map(x -> colorscheme.colors[trunc(Int,x.val * (k-1))+1], img)
end
```

## Results

```julia
using Colors
rainbow = [convert(RGB, HSV(i * 240, 1.0, 1.0)) for i in range(0,stop=1,length=100)]
```

```julia
imresize(which_blair_test(ColorScheme(rainbow)), ratio=0.2)
```

```julia
imresize(which_blair_test(ColorSchemes.viridis), ratio=0.2)
```

## Discussion

Subjective, but most people think the viridis image looks more natural than the rainbow image. We conclude that Viridis is better than Rainbow for rendering continuous scalar data.
