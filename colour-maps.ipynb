## Recreate colour maps from Rogwitz and Kalvin (2001)

```julia
heated_body = [HSV(i * 60, 1.0, i) for i in range(0,stop=1,length=100)]
```

```julia
hsv_grayscale = [HSV(0, 0, v) for v in range(0,stop=1,length=100)]
```

```julia
hsv_increasing_saturation = [HSV(300, s, 1.0) for s in range(0,stop=1,length=100)]
```

```julia
hsv_decreasing_saturation = reverse(hsv_increasing_saturation)
```

```julia
lab_grayscale = [Lab(l,0,0) for l in range(0,stop=100,length=100)]
```

```julia
rainbow = [HSV(i * 240, 1.0, 1.0) for i in range(0,stop=1,length=100)]
```

```julia
using Interpolations

function interp1(X, V, Xq)
    knots = (X,)
    itp = interpolate(knots, V, Gridded(Linear()))
    itp[Xq]
end

linspace(x,y,z) = range(x, stop=y,length=z)
```

```julia
# Load the colours from the digitised colour map
x = unique(Images.load("data/raw/isoluminant-rainbow-sample.png")[1,:])

# Convert to CIELab
x = map( x -> Lab(RGB(x.r, x.g, x.b)), x)

# Find the mean lightness and use it to ensure isoluminance
m = mean([x.l for x in x])
x = map(x -> Lab(m, x.a, x.b), x)

# Interpolate to get 100 colours
idx1 = linspace(1,100, length(x))
idx2 = collect(1:100)
x = [[x.l x.a x.b] for x in x]
x = interp1(idx1, x, idx2)
isoluminant_rainbow = [Lab(x[1],x[2],x[3]) for x in x]
```

```julia
x = unique(load("data/raw/lab-isoluminant-saturation-sample.png")[1,:])
x = map( x -> Lab(RGB(x.r, x.g, x.b)), x)
m = mean([x.l for x in x])
x = map(x -> Lab(m, x.a, x.b), x)
idx1 = linspace(1,100, length(x))
idx2 = collect(1:100)
x = [[x.l x.a x.b] for x in x]
x = interp1(idx1, x, idx2)
isoluminant_sat = [Lab(x[1],x[2],x[3]) for x in x]
```
